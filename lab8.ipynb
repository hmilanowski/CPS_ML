```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

from sklearn.datasets import make_moons
import seaborn as sns

from sklearn.datasets.samples_generator import make_blobs 
 
help(make_blobs)
```

# K-means clustering - example I

Ex.2 Please call make_blob and generate a dataset with 120 samples, 2 features, 4 centres (clusters), and cluster_std=0.40

```python
xBlob, yBlob = make_blobs(n_samples=120, centers=4, n_features=2,
                          cluster_std=0.4, shuffle=True)
```

Ex.3 Analyse the x and y shape.

```python
print(xBlob.shape)
print(yBlob.shape)
```

Ex.4 Plot the data samples in X to see the clusters.

```python
plt.scatter(xBlob[:, 0], xBlob[:, 1], c='blue', marker='.')
plt.show()
```

Ex.5 Run the implemented KMeans function K-Means, fit the model and predict the outcome.

```python
km = KMeans(n_clusters=4, init='random')
km.fit_predict(xBlob, yBlob)
```

Ex.6 Plot the results in order to see whether the clustering worked.

```python
plt.figure(figsize=(6, 5))
plt.scatter(xBlob[:, 0], xBlob[:, 1], s=70, c=km.labels_)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')
```

Ex.7 Calculate inertia Inertia_ or Dunn index Package jgmcvi.

```python
print(km.inertia_)
```

# Elbow method

Ex.1 Run K-Means for a range of clusters using a for loop and collecting the distortions (inertia) into a list.
Ex.2 Plot the distortions of K-Means.

```python
def elbow_method(x_data, y_data):
    inertia_list = []
    k_list = range(1, 10)
    for k in k_list:
        temp = KMeans(n_clusters=k)
        temp.fit_predict(x_data, y_data)
        inertia_list.append(temp.inertia_)
        
    plt.plot(k_list, inertia_list)
    plt.show()
```

```python
blob_means = elbow_method(xBlob, yBlob)
```

# Please check whether the K-Means algorithm will correctly cluster the data: 

```python
xMoon, yMoon = make_moons(1000, noise=.05, random_state=0)
X_moon = pd.DataFrame(xMoon, columns=['f1', 'f2'])

kmeans = KMeans(n_clusters=2, random_state=0).fit(xMoon)
y_km = kmeans.predict(xMoon)
centers = kmeans.cluster_centers_

X_moon['k_means'] = y_km
sns.lmplot(data=X_moon, x='f1', y='f2', fit_reg=False, hue = 'k_means', palette = ['#eb6c6a', '#6aeb6c']).set(title='Algorytm k-srednich')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha=0.5)
plt.show()
```
