```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

irisRaw = load_iris()
 
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
irisPetalLength = iris[iris.columns[2]]
mxidx = irisPetalLength.idxmax()
mx = irisPetalLength[mxidx]
mnidx = irisPetalLength.idxmin()
mn = irisPetalLength[mnidx]

print('Maximum peta length in dataset {} at index {}.'.format(mx, mxidx))
print('Minimum peta length in dataset {} at index {}.'.format(mn, mnidx))
```

```python
irisSepalLength = iris['sepal length (cm)']
irisSepalMean = irisSepalLength.mean()

irisCleaned = iris[iris['sepal length (cm)'] > irisSepalMean]

irisCleaned
```

```python
c0 = iris[iris.target == 0]
c1 = iris[iris.target == 1]
c2 = iris[iris.target == 2]

c0.drop(columns=['target'], inplace=True)
c1.drop(columns=['target'], inplace=True)
c2.drop(columns=['target'], inplace=True)

c0Hist = c0.plot.hist(grid=True, bins=20, rwidth=0.9)
c1Hist = c1.plot.hist(grid=True, bins=20, rwidth=0.9)
c2Hist = c2.plot.hist(grid=True, bins=20, rwidth=0.9)

plt.grid(axis='y', alpha=0.75)
plt.show()
```
